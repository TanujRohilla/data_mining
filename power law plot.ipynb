```python
import matplotlib.pyplot as plt
x=[]
y=[]
for i in range(1,100):
    x.append(i)
    y.append(pow(i,-0.5))

plt.xlabel('Kin')
plt.ylabel('Pkin')
plt.suptitle('For gamma = 0.5')
plt.plot(x,y)
plt.show()
```
